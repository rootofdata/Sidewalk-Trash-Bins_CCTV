# 쓰레기 배출량

```python
x = ['2018','2019','2020','2021 예상']
y2 = np.array([2045,2116,2254,2347])
dt=pd.DataFrame({'년도':['2018년','2019년','2020년','2021년(예상)'],'쓰레기 배출량': [2045,2116,2254,2347]})
dt
```

```python
# 생활폐기물 발생량 시각화
fig = plt.figure(figsize=(12,6))
ax=sns.barplot(x="년도",y="쓰레기 배출량",data=dt,color="#08298A")

bg_color="#f1f1f1"
fig.patch.set_facecolor(bg_color)
fig.set_facecolor(bg_color)

for p in ax.patches:
    ax.annotate(format(p.get_height(),'.0f'),
               (p.get_x()+p.get_width()/2.,
               p.get_height()),
               ha='center',va='center',
               size=12,color='grey',
               xytext=(0,14),
               textcoords='offset points')
plt.xlabel("년도",size=14)
plt.ylabel("폐기물량(만톤)",size=14)
plt.title("생활 폐기물 발생량",size=20,weight='bold',pad='13')
plt.ylim(2000,2400,100) 
plt.box(False)
plt.grid(True,axis='y',alpha=10,linestyle='--')
plt.show()
```

```python
import numpy

# 2. 데이터 준비
x = ['2020년 1월','2020년 2월','2020년 3월']
y2 = np.array([5349,5355,5521])
y1 = np.array([809,839,868])

# 3. 그래프 그리기
fig,ax1=plt.subplots(figsize=(6,5.5))

#선그래프
ax1.plot(x, y1, '-s', color='red', markersize=5, linewidth=5, alpha=0.7, label='플라스틱류')
ax1.set_ylim(0, 1000)
ax1.set_xlabel('월')
ax1.set_ylabel('플라스틱류')
ax1.tick_params(axis='both', direction='in')

for i in range(len(x)):
    height = y1[i]
    plt.text(x[i], height + 0.25, '%.0f' %height, ha='center', va='bottom', size = 12)
#박스그래프
ax2 = ax1.twinx()
ax2.bar(x, y2, color='blue', label='전체', alpha=0.7, width=0.2)
ax2.set_ylim(5000,6000)
ax2.set_ylabel('배출량')
ax2.tick_params(axis='y', direction='in')

plt.title('코로나 19 이후 생활폐기물 재활용 쓰레기 품목별 통계')
ax1.legend(loc='upper right')
ax2.legend(loc='upper left')
plt.show()
```
